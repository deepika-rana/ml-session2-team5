```python
import pandas as pd

df=pd.read_csv('Pytorch_Community_Scraped_Data_7059posts.csv')
df
```

```python
df.info()
```

```python
df['post_content'] = df['Topic Title'] + ' ' + df['Leading Post'] + ' ' + df['Post Replies']
df['post_content']
```

```python
df['word_count'] = df['post_content'].apply(lambda x: len(str(x).split(" ")))
df['char_count'] = df['post_content'].str.len()
df
```

```python
df[['post_content','word_count','char_count']]
```

```python
import re
import string
# Creating a 'clean_text' function
def clean_text(text):
    #Make text lowercase, remove punctuation and remove words containing numbers.
    text = re.sub("[\(\[].*?[\)\]]", "", text)
    text = re.sub('<.*?>','', text)
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)#remove punctuation
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub(' — ', '', text)
    return text
```

```python
df['clean_post_content'] = df['post_content'].apply(lambda x: clean_text(str(x)))
df[['clean_post_content','word_count','char_count']]
```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
df['stop_words'] = df['clean_post_content'].apply(lambda x:([x for x in x.split() if x in stop_words]))
df[['clean_post_content','word_count','char_count','stop_words']]
```

```python
def remove_stop_words(content):
    stop_words = stopwords.words('english')
    content = ' '.join(x for x in str(content).split() if x not in stop_words)
    return content
```

```python
df['clean_post_content_w/o_stop_words'] = df['clean_post_content'].apply(lambda x: remove_stop_words(str(x)))
df['clean_post_content_w/o_stop_words']
```

```python
df
```

```python
from nltk.tokenize import word_tokenize
df['Tokenization'] = df['clean_post_content_w/o_stop_words'].apply(word_tokenize)
df['Tokenization']
```

```python
import nltk
nltk.download('all')
```

```python
df['pos_tags'] = df['Tokenization'].apply(nltk.tag.pos_tag)
df['pos_tags'] 
```

```python
from nltk.corpus import wordnet

def get_wordnet_pos(pos_tag):
    #print(pos_tag[0])
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    
    for i in tag_dict:
        if i == pos_tag[0]:
            return tag_dict[i]
        else:
            return wordnet.NOUN

df['wordnet_pos'] = df['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
```

```python
df['wordnet_pos']
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
df['lamentation'] = df['wordnet_pos'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word,tag) for (word,tag) in x]))
df['lamentation']
```
